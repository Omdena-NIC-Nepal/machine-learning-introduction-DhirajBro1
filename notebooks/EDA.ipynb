```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
import sklearn
import seaborn as sns
from matplotlib import rcParams
```

#### Load the dataset

```python

data = pd.read_csv('../data/boston_housing.csv')  
data.head()
```

##### Explore data

```python
data.info()
data.describe()
```

```python
#Renaming the MEDV into price
data = data.rename(columns={"medv": "price"})
```

```python
# Print the column names of your dataset to check if 'RM' is included
print(data.columns)
```

##### Visualize relationsip

```python
sns.regplot(y="price", x="rm", data=data, fit_reg = True)
plt.title("Relationship between RM and Price")
```

```python
sns.regplot(y="price", x="lstat", data=data, fit_reg = True)
plt.title("Relationship between LSTAT and Price")
```

```python
sns.regplot(y="price", x="nox", data=data, fit_reg = True)
plt.title("Relationship between NOX and Price")
```

```python
sns.regplot(y="price", x="dis", data=data, fit_reg = True)
plt.title("Relationship between DIS and Price")
```

```python
sns.pairplot(data, x_vars=['rm', 'lstat', 'ptratio'], y_vars='price', kind='scatter')

# Correlation Heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm')
```
